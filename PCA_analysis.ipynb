```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data_folder = 'data'
storage_cond_data = os.path.join(data_folder, 'Study 2 - Storage conditions -MDataSci 2021.csv')

data = pd.read_csv(storage_cond_data)
```

```python
data.columns
```

```python
X = data.iloc[:,2:]
X_normalized = normalize(X)

plt.matshow(np.corrcoef(X_normalized, rowvar=False))
plt.show()
```

```python
pca = PCA()
pca.fit(X_normalized)
```

```python
pca.transform(X_normalized)
```
