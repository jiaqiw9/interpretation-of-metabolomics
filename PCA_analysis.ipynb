```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data_folder = 'data'
storage_cond_data = os.path.join(data_folder, 'Study 2 - Storage conditions -MDataSci 2021.csv')

data = pd.read_csv(storage_cond_data)
```

```python
data.columns
```

```python
data
```

```python
X = data.iloc[:,2:]
X_normalized = normalize(X)

figure = plt.figure()
axes = figure.add_subplot(111)

caxes = axes.matshow(np.corrcoef(X_normalized, rowvar=False))
figure.colorbar(caxes)
plt.show()
```

```python
X_normalized[:,:5]
```

```python
pca = PCA(n_components=4)
pca.fit(X_normalized)
```

```python
pca_data = pca.transform(X_normalized)
pca_df = pd.DataFrame(pca_data, columns = ['Feature' + str(i) for i in range(1,5)])
pca_df.to_csv(os.path.join(data_folder, 'PCA_features.csv'))
```
